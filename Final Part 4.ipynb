It is clear that method 2 costs the most time. The reason is that DataFrame follows column-major order, so every time we circularly assign the value to elements by row, we jump to another unit of this DataFrame once which definitely costs much more time than successively assigning the value to elements in the same unit.

```python
# Chunk 1
A = np.array([i for i in range(10000)]).reshape(100, 100)
for i in A:
    A[i] = np.array([i for i in range(100)])
A
# Chunk 2
A = np.array([i for i in range(10000)]).reshape(100, 100)
A = pd.DataFrame(A)
for i in A:
    A[i] = np.array([i for i in range(100)])
A
```

The output of Chunk 1 and Chunk 2 is clearly not the same one, because a numpy.array follows row-major order,which means the base unit of a array is a row instead of a column, while a DataFrame is in column-major order.
